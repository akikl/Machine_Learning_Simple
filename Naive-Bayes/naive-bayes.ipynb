A model is built to classify wines into 3 categories. Load the dataset and split it into test and train.

```python
for cols in num_cols:
    print(cols)
    print('Skew :', round(df[cols].skew(), 2))
    plt.figure(figsize=(15,4))
    plt.subplot(1,2,1)
    sns.distplot(df[cols])
    plt.subplot(1,2,2)
    sns.boxplot(x=df[cols])
    plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(df.corr(),annot=True)
```

```python
X=df.drop('wine_type',axis=1)
```

```python
y=df.wine_type
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
mod1=GaussianNB()
```

```python
mod2=MultinomialNB()
```

```python
accuracy=[]
```

```python
for model in mod1,mod2:
    clf=model
    clf.fit(X_train,y_train)
    pred=clf.predict(X_test)
    accuracy.append(accuracy_score(y_test,pred))
```

```python
accuracy
```

Gaussian Naive Bayes gives Better accuracy than Multinomial Naive Bayes model
